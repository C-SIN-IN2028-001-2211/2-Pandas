# Combinaci√≥n

En ocasiones requerimos combinar informaci√≥n de diferentes Dataframes

```python
import pandas as pd

canadian_youtube = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/CAvideos.csv")
british_youtube = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/GBvideos.csv")
```

```python
pd.concat([canadian_youtube, british_youtube])
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])
```
